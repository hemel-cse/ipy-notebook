```python
import mechanize
import cookielib
from bs4 import BeautifulSoup as bs
import html2text
from time import sleep
import re

br = mechanize.Browser()

cj = cookielib.LWPCookieJar()

br.set_cookiejar(cj)

br.set_handle_equiv(True)

br.set_handle_gzip(True)

br.set_handle_redirect(True)

br.set_handle_referer(True)

br.set_handle_robots(False)

br.set_handle_refresh(mechanize._http.HTTPRefreshProcessor(), max_time=1)

br.addheaders = [('User-agent', 'Chrome')]

bsite = br.open('http://www.datacenter9.com/datacenters/united-kingdom').read()
```

```python
sp = bs(bsite)

table = sp.find('table', {'class': 'datacenter_tab'});

rows = table.findAll('tr')


li = []
final_list = []
in_list = []
```

```python
for tr in rows:
    for td in tr.findAll("a", {"href": True}):
        li.append("http://www.datacenter9.com"+td["href"])
for tr in rows:
    for td in tr.findAll("a", {"href": True}):
        in_list.append(td.text)
```

```python
def linfunc(lnk):
    bbb = br.open(lnk).read()
    nnn = bs(bbb)
    tablech = nnn.find('table', {'class': 'datacenter_tab_light'});
    rowsch = tablech.findAll('tr')
    for tr in rowsch:
        for td in tr.findAll("a", {"href": True}):
            final_list.append(td.text)
    return

for i in li:
    linfunc(i)
    sleep(1)
```
